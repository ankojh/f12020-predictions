```python
def getDriverPracticeData(df, venue, driver, practiceSession):
    pracData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    pracData = pracData.loc[pracData['tabs'] == practiceSession][['Time', 'Gap', 'Laps', 'Pos']]
    if(pracData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, practiceSession+'-time'] = pracData.iloc[0]['Time']
    df.loc[df['drivers'] == driver, practiceSession+'-gap']= pracData.iloc[0]['Gap']
    df.loc[df['drivers'] == driver, practiceSession+'-laps'] = pracData.iloc[0]['Laps']
    df.loc[df['drivers'] == driver, practiceSession+'-pos'] = pracData.iloc[0]['Pos']
    
```

```python
def getDriverQualiData(df, venue, driver, qualiSession):
    qualiData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    qualiData = qualiData.loc[qualiData['tabs'] == 'Qualifying']
    if(qualiData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, qualiSession+'-pos'] = qualiData.iloc[0]['Pos']
    df.loc[df['drivers'] == driver, qualiSession+'-laps'] = qualiData.iloc[0]['Laps']
```

```python
def getDriverGridData(df, venue, driver):
    gridData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    gridData = gridData.loc[gridData['tabs'] == 'Starting grid']
    if(gridData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, 'grid-pos'] = gridData.iloc[0]['Pos']
```

```python
def getDriverRaceData(df, venue, driver):
    raceData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    raceData = raceData.loc[raceData['tabs'] == 'Qualifying']
    if(raceData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, 'race-pos'] = raceData.iloc[0]['Pos']
```

```python
def getDriverFastestLapData(df, venue, driver):
    flData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    flData = flData.loc[flData['tabs'] == 'Fastest laps']
    if(flData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, 'fastestlap-pos'] = flData.iloc[0]['Pos']
    df.loc[df['drivers'] == driver, 'fastestlap-time'] = flData.iloc[0]['Time']
```

```python
def getDriverPitStopData(df, venue, driver):
    pitData = df_raw[(df_raw['venue'] == venue) & (df_raw['Num'] == driver)]
    pitData = pitData.loc[pitData['tabs'] == 'Pit stop summary']
    if(pitData.shape[0] == 0):
        return
    df.loc[df['drivers'] == driver, 'pit-time'] = pitData.iloc[0]['Time']
```

```python
def getVenueDF(venue):
    vdf = pd.DataFrame()
    vdf['drivers'] = drivers
    vdf = vdf.assign(venue=venue)
    for driver in drivers:
#         getDriverPracticeData(vdf ,venue, driver, 'Practice 1');
        getDriverPracticeData(vdf ,venue, driver, 'Practice 2');
        getDriverPracticeData(vdf ,venue, driver, 'Practice 3');
        getDriverQualiData(vdf ,venue, driver, 'Quali');
#         getDriverQualiData(venue, driver, 'Q1');
#         getDriverQualiData(venue, driver, 'Q1');
#         getDriverQualiData(venue, driver, 'Q1');
        getDriverFastestLapData(vdf,venue, driver);
        getDriverPitStopData(vdf,venue, driver);
        getDriverGridData(vdf,venue, driver);
        getDriverRaceData(vdf,venue, driver);
    return vdf
```

```python
venueDFs = []
venues = ['Spain', 'Great Britain', 'Styria', '70th Anniversary', 'Italy',
       'Tuscany', 'Belgium', 'Hungary']

for venue in venues:
    venueDF = getVenueDF(venue);
    venueDFs.append(venueDF)

cdf = pd.concat(venueDFs)
cdf
```
